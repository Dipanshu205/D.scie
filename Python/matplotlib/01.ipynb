```python
x = ['mon','tues','wed','thrus','fri']
y = [20,22,28,28,25]
plt.plot(x,y)
plt.title('Bakery Sales this week')
plt.xlabel('Days of the Week')
plt.ylabel('Sales per day')
plt.show()
```

```python
plt.plot(x,y, marker='s', linestyle='--', color='g', linewidth=2, label='Sales data')
plt.title('Bakery Sales this week')
plt.xlabel('Days of the Week')
plt.ylabel('Sales per day')
plt.legend(loc='upper left', fontsize=8)       ######
plt.grid(color='gray',linestyle= ':', linewidth='2')
plt.xlim(0,4)
plt.ylim(0,40)
plt.xticks(['mon','tue','wed','thrus','fri'], ['d1','d2','d3','d4','d5'])
plt.show()
```

## Object oriented api

```python
# plt.bar(x, height, color='name', width='value', label='label name')
product = ['a','b','c','d']
sales = [1000, 1500, 800, 1200]

plt.bar(product,sales, color= 'orange',label = 'sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()
plt.show()


plt.barh(product,sales, color= 'orange',label = 'sales 2025')
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sale Comparison')
plt.legend()
plt.show()
```

```python
# plt.pie(values, labels = label_list, colors=clor_list, autopct= '%1.1f%')

regions = ['north','south','east','west']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue,labels =regions, autopct='%1.1f%%', colors=['gold','skyblue','lightgreen','coral'])
plt.title('Revenue contribution by region')
plt.show()
```

```python
# Histogram
# plt.hist[data, bins= numberofbins, color='colorname',edgecolor='black']

scores = [54,34,65,24,57,43,23,56,34,23,65,23,75,78,54,89]
plt.hist(scores, bins=5, color='lavender',edgecolor='gray')
plt.xlabel("score range")
plt.ylabel('number of students')
plt.title('score distribution of syudents')
plt.show()
```

```python
# plt.scatter(x,y, color='colornames', marker='marker style', lbel='label name')

hours_studied = [1,2,3,4,5,6,7,8]
exam_scores = [56,87,23,54,23,45,23,34]

# scatter plot
plt.scatter(hours_studied,exam_scores, color='green', marker='o', label='student data')
plt.xlabel("hours studied")
plt.ylabel("exam score")
plt.title("relation bw study time and exam score")
plt.legend()
plt.grid(True) 
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,15,25]

plt.subplot(1,2,1) # 1 row, 2 colomn , 1st subplot
plt.plot(x,y)
plt.title('Line chart')

plt.subplot(1,2,2) # 1 row, 2 colomn , 2nd subplot
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
plt.show()
```

```python

# fig, ax = plt.subplot(nrowa, ncols, figsize=(width, height))
fig, ax = plt.subplots( 1, 2, figsize = (10,5))

x=[1,2,3,4]
y=[10,20,15,25]

ax[0].plot(x,y)
ax[0].set_title('Line plot')

ax[1].bar(x,y, color='green')
ax[1].set_title('Bar Chart')

fig.suptitle('Coomparison of line and Bar charts')

plt.tight_layout()
plt.show()
```
